# 🏠 Swiss Mortgage Simulator — 2 Tranches
Assumes **20% down payment** and **5% notary fees**.  
Monthly interest charged on outstanding principal (*Festhypothek*).  
Principal amortised at **1% p.a.** of original mortgage (pro-rata per tranche).  
**Bonus Vert**: waives Tranche 1 interest for the first 12 months.

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import IntText, FloatText, IntSlider, Checkbox, HBox, VBox, HTML
import ipywidgets as widgets

# ── colour palette ───────────────────────────────────────────────
C1      = '#2563EB'   # tranche 1  – blue
C2      = '#16A34A'   # tranche 2  – green
C_AMORT = '#F59E0B'   # amortisation – amber
C_TOTAL = '#64748B'   # total – slate
C_SAVE  = '#7C3AED'   # savings – violet
C_BONUS = '#10B981'   # bonus vert highlight
BG      = '#F8FAFC'
GRID    = '#E2E8F0'

# ── widgets ──────────────────────────────────────────────────────
w_price  = IntText(value=900_000, description='Property (CHF)',
                   style={'description_width':'130px'}, layout=widgets.Layout(width='280px'))
w_rate1  = FloatText(value=1.8,  description='Rate T1 (%)',
                     style={'description_width':'130px'}, layout=widgets.Layout(width='240px'), step=0.05)
w_rate2  = FloatText(value=2.2,  description='Rate T2 (%)',
                     style={'description_width':'130px'}, layout=widgets.Layout(width='240px'), step=0.05)
w_years1 = IntSlider(value=5, min=2, max=10, step=1, description='Years T1',
                     style={'description_width':'100px'}, layout=widgets.Layout(width='350px'))
w_years2 = IntSlider(value=5, min=2, max=10, step=1, description='Years T2',
                     style={'description_width':'100px'}, layout=widgets.Layout(width='350px'))
w_split  = FloatText(value=50.0, description='T1 share (%)',
                     style={'description_width':'130px'}, layout=widgets.Layout(width='240px'), step=5.0)
w_bonus  = Checkbox(value=False, description='🌿 Bonus Vert  (waive T1 interest – Year 1)',
                    style={'description_width':'initial'}, layout=widgets.Layout(width='380px'))

header = HTML('<h3 style="font-family:Georgia,serif;margin-bottom:4px">🏔 Swiss Mortgage Simulator</h3>')
note   = HTML(
    '<p style="color:#64748b;font-size:0.85em;font-family:Georgia,serif">' +
    'Down payment 20% · Notary fees 5% · Festhypothek · '
    'Amortisation 1% p.a. of original mortgage (pro-rata per tranche)</p>'
)

controls = VBox([
    header, note,
    HBox([w_price]),
    HBox([w_rate1, w_rate2, w_split]),
    HBox([w_years1]),
    HBox([w_years2]),
    HBox([w_bonus]),
])

out = widgets.Output()

# ── core simulation ───────────────────────────────────────────────
def simulate(price, rate1, rate2, years1, years2, split_pct, bonus_vert):
    DOWN       = 0.20
    NOTARY     = 0.05
    AMORT_RATE = 0.01

    total_cost   = price * (1 + NOTARY)
    down_payment = price * DOWN
    mortgage     = total_cost - down_payment

    split = np.clip(split_pct, 1, 99) / 100
    p1_0  = mortgage * split
    p2_0  = mortgage * (1 - split)

    monthly_amort1 = mortgage * AMORT_RATE * split       / 12
    monthly_amort2 = mortgage * AMORT_RATE * (1 - split) / 12

    months = 120
    m1 = years1 * 12
    m2 = years2 * 12
    r1 = rate1 / 100 / 12
    r2 = rate2 / 100 / 12

    def run(apply_bonus):
        p1 = np.zeros(months + 1);  p1[0] = p1_0
        p2 = np.zeros(months + 1);  p2[0] = p2_0
        ic1 = np.zeros(months)
        ic2 = np.zeros(months)
        am  = np.zeros(months)
        for i in range(months):
            v1 = max(p1[i], 0.0)
            v2 = max(p2[i], 0.0)
            eff_r1 = r1 if (i // m1) == 0 else r2
            eff_r2 = r2
            # Bonus Vert: zero T1 interest for first 12 months
            if apply_bonus and i < 12:
                ic1[i] = 0.0
            else:
                ic1[i] = v1 * eff_r1
            ic2[i] = v2 * eff_r2
            a1 = min(monthly_amort1, v1)
            a2 = min(monthly_amort2, v2)
            am[i] = a1 + a2
            p1[i+1] = v1 - a1
            p2[i+1] = v2 - a2
        return ic1, ic2, am, p1[:months], p2[:months]

    ic1, ic2, am, pr1, pr2 = run(bonus_vert)
    ic1_base, ic2_base, am_base, pr1_base, pr2_base = run(False)

    total     = ic1 + ic2 + am
    total_base= ic1_base + ic2_base + am_base
    savings   = total_base - total          # monthly saving vs no-bonus
    cum_save  = np.cumsum(savings)

    return dict(
        mortgage       = mortgage,
        down           = down_payment,
        notary         = price * NOTARY,
        p1_0=p1_0, p2_0=p2_0,
        monthly_amort  = mortgage * AMORT_RATE / 12,
        ic1=ic1, ic2=ic2, am=am,
        total=total,
        principal1=pr1, principal2=pr2,
        savings=savings, cum_save=cum_save,
        total_bonus_saving = cum_save[-1],
        months=months, m1=m1, m2=m2,
    )

# ── plot ──────────────────────────────────────────────────────────
def update(*args):
    out.clear_output(wait=True)
    price     = w_price.value
    rate1     = w_rate1.value
    rate2     = w_rate2.value
    years1    = w_years1.value
    years2    = w_years2.value
    split_pct = w_split.value
    bonus     = w_bonus.value

    if price <= 0 or rate1 <= 0 or rate2 <= 0:
        return

    d          = simulate(price, rate1, rate2, years1, years2, split_pct, bonus)
    time_years = np.arange(d['months']) / 12
    xticks     = np.arange(0, 11, 1)

    ncols = 3 if bonus else 2

    with out:
        fig, axes = plt.subplots(1, ncols, figsize=(7*ncols, 5.5), facecolor=BG)
        if ncols == 2:
            axes = list(axes)
        fig.subplots_adjust(wspace=0.38)

        bonus_tag = '  🌿 Bonus Vert ACTIVE – T1 interest waived Year 1' if bonus else ''
        fig.suptitle(
            f"  Purchase: CHF {price:,.0f}  │  Notary: CHF {d['notary']:,.0f}  │  "
            f"Down: CHF {d['down']:,.0f}  │  Mortgage: CHF {d['mortgage']:,.0f}  │  "
            f"T1: CHF {d['p1_0']:,.0f} @ {rate1:.2f}%  │  "
            f"T2: CHF {d['p2_0']:,.0f} @ {rate2:.2f}%  │  "
            f"Amort: CHF {d['monthly_amort']:,.0f}/mo{bonus_tag}",
            fontsize=8.0, color='#475569', fontfamily='monospace', y=1.02
        )

        def add_renewal_markers(ax, ymax):
            for k in range(1, 11):
                t = k * years1
                if 0 < t < 10:
                    ax.axvline(t, color=C1, linewidth=1, linestyle='--', alpha=0.5)
                    ax.text(t + 0.06, ymax * 0.97, 'T1↺', fontsize=7, color=C1, va='top')
            for k in range(1, 11):
                t = k * years2
                if 0 < t < 10:
                    ax.axvline(t, color=C2, linewidth=1, linestyle=':', alpha=0.5)
                    ax.text(t + 0.06, ymax * 0.85, 'T2↺', fontsize=7, color=C2, va='top')

        # ── Plot 1: stacked monthly cost ──────────────────────────
        ax1 = axes[0]
        ax1.set_facecolor(BG)
        ax1.stackplot(
            time_years,
            d['ic1'], d['ic2'], d['am'],
            labels=['Interest T1', 'Interest T2', 'Amortisation (1% p.a.)'],
            colors=[C1, C2, C_AMORT], alpha=0.75
        )
        ax1.plot(time_years, d['total'], color=C_TOTAL,
                 linewidth=1.8, linestyle='--', label='Total')
        if bonus:
            ax1.axvspan(0, 1, color=C_BONUS, alpha=0.08, zorder=0)
            ax1.text(0.03, d['total'].max() * 1.05, '🌿 Bonus Vert',
                     fontsize=8, color=C_BONUS, va='bottom')
        ymax1 = d['total'].max() * 1.15
        add_renewal_markers(ax1, ymax1)
        ax1.set_xlim(0, 10); ax1.set_ylim(0, ymax1)
        ax1.set_xticks(xticks)
        ax1.set_xlabel('Year', fontsize=10)
        ax1.set_ylabel('CHF / month', fontsize=10)
        ax1.set_title('Monthly Cost Breakdown', fontsize=12, fontweight='bold', pad=10)
        ax1.legend(fontsize=8.5, framealpha=0.6, loc='upper right')
        ax1.grid(True, color=GRID, linewidth=0.8, zorder=0)
        ax1.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x:,.0f}'))

        # ── Plot 2: outstanding principal ─────────────────────────
        ax2 = axes[1]
        ax2.set_facecolor(BG)
        total_p = (d['principal1'] + d['principal2']) / 1000
        ax2.fill_between(time_years, d['principal1'] / 1000, alpha=0.15, color=C1)
        ax2.fill_between(time_years, d['principal2'] / 1000, alpha=0.15, color=C2)
        ax2.plot(time_years, d['principal1'] / 1000, color=C1, linewidth=2.5,
                 label=f'Tranche 1  ({rate1:.2f}%,  {years1}yr)')
        ax2.plot(time_years, d['principal2'] / 1000, color=C2, linewidth=2.5,
                 label=f'Tranche 2  ({rate2:.2f}%,  {years2}yr)')
        ax2.plot(time_years, total_p, color=C_TOTAL, linewidth=1.8,
                 linestyle='--', label='Total')
        ax2.set_xlim(0, 10); ax2.set_xticks(xticks)
        ax2.set_xlabel('Year', fontsize=10)
        ax2.set_ylabel('Outstanding Principal (CHF k)', fontsize=10)
        ax2.set_title('Outstanding Principal per Tranche', fontsize=12, fontweight='bold', pad=10)
        ax2.legend(fontsize=9, framealpha=0.5)
        ax2.grid(True, color=GRID, linewidth=0.8)
        ax2.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x:,.0f}k'))

        # ── Plot 3: Bonus Vert savings (only when active) ─────────
        if bonus:
            ax3 = axes[2]
            ax3.set_facecolor(BG)

            # Monthly saving bars (only non-zero in first 12 months)
            bar_colors = [C_BONUS if s > 0 else '#E2E8F0' for s in d['savings']]
            ax3.bar(time_years, d['savings'], width=1/12,
                    color=bar_colors, alpha=0.6, label='Monthly saving')

            # Cumulative savings line on twin axis
            ax3b = ax3.twinx()
            ax3b.plot(time_years, d['cum_save'], color=C_SAVE, linewidth=2.5,
                      label='Cumulative saving')
            ax3b.fill_between(time_years, d['cum_save'], alpha=0.10, color=C_SAVE)
            ax3b.set_ylabel('Cumulative Saving (CHF)', fontsize=10, color=C_SAVE)
            ax3b.tick_params(axis='y', colors=C_SAVE)
            ax3b.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x:,.0f}'))
            ax3b.set_ylim(0, d['cum_save'].max() * 1.35)

            # Annotate total saving
            total_saved = d['total_bonus_saving']
            ax3b.annotate(
                f"Total saved:\nCHF {total_saved:,.0f}",
                xy=(1.0, total_saved),
                xytext=(3.5, total_saved * 1.15),
                fontsize=9, color=C_SAVE, fontweight='bold',
                arrowprops=dict(arrowstyle='->', color=C_SAVE, lw=1.5)
            )

            # Highlight the bonus window
            ax3.axvspan(0, 1, color=C_BONUS, alpha=0.07, zorder=0)
            ax3.axvline(1, color=C_BONUS, linewidth=1.2, linestyle='--', alpha=0.7)
            ax3.text(1.05, ax3.get_ylim()[1] if ax3.get_ylim()[1] > 0 else 1,
                     'End of\nBonus Vert', fontsize=7.5, color=C_BONUS, va='top')

            ax3.set_xlim(0, 10); ax3.set_xticks(xticks)
            ax3.set_xlabel('Year', fontsize=10)
            ax3.set_ylabel('Monthly Saving (CHF)', fontsize=10)
            ax3.set_title('🌿 Bonus Vert — Savings vs Standard Mortgage',
                          fontsize=12, fontweight='bold', pad=10, color='#065F46')
            ax3.grid(True, color=GRID, linewidth=0.8, zorder=0)
            ax3.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{x:,.0f}'))

            # combined legend
            h1, l1 = ax3.get_legend_handles_labels()
            h2, l2 = ax3b.get_legend_handles_labels()
            ax3.legend(h1 + h2, l1 + l2, fontsize=8.5, framealpha=0.6, loc='center right')
            ax3b.spines['right'].set_edgecolor(C_SAVE)

        for ax in ([ax1, ax2] + ([axes[2]] if bonus else [])):
            for spine in ax.spines.values():
                if spine != (ax3b.spines['right'] if bonus else None):
                    spine.set_edgecolor(GRID)

        plt.tight_layout()
        plt.show()

# ── wire up observers ─────────────────────────────────────────────
for w in [w_price, w_rate1, w_rate2, w_years1, w_years2, w_split, w_bonus]:
    w.observe(update, names='value')

display(controls, out)
update()
```
